```python
import pandas as pd

#read the dataset
#读取数据集
file_path = r'C:\Users\yuentolong\Dataset .csv'
data = pd.read_csv(file_path,encoding = 'latin-1') #Use the 'latin-1' encoding to read data correctly #使用latin-1编码正确读取数据

#The city with the most restaurants
#最多餐厅的城市
city_counts = data['City'].value_counts()

#Print the result
#输出结果
print(f"The city with the most restaurants is:{city_counts.head(1).to_string(index = True,header = False)}") #Use the 'to_string' function to disable index  #使用to_string函数禁用索引
print(f"最多餐厅的城市：{city_counts.head(1).to_string(index = True,header = False)}")
```

```python
#Calculate the average rating for restaurants for each city
#计算各个城市餐厅的平均评分
dataGroup =data[['City','Aggregate rating']].groupby('City')
Rating_sum = dataGroup.sum()

Average_rating = data.groupby('City')['Aggregate rating'].mean().round(3)
Average_rating.sort_values(ascending = False)

#         Average_rating = Rating_sum['Aggregate rating'] / city_counts
#         Average_rating.sort_values(ascending = False)   
#      #Inferior method: calculating average rating directly via division                             劣等方法：通过除法直接计算平均评分
#      #(commented out — this fails with NaN values or zero count; use .mean() instead for safety)  （已注释 —— 当存在 NaN 值或计数为零时会计算失败；为了数据安全使用 .mean())

#Print the city with the highest average rating
#输出最高平均分的城市
print(f"The city with the highest average rating:{Average_rating.head(1).to_string(index = True,header = False)}")
```
