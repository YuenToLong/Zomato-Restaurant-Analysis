```python
import pandas as pd

#read the dataset
#è¯»åæ°æ®é
file_path = r'C:\Users\yuentolong\Dataset .csv'
df = pd.read_csv(file_path,encoding = 'latin-1')

df.info()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#The number of bins is set to ân to balance detail and stability in the distribution. ç®±å­çæ°éè®¾ç½®ä¸ºânï¼ä»¥å¹³è¡¡åå¸çç»èåç¨³å®æ§ã
#Different binning strategies may yield different results, so interpretability is important. ä¸åçåç®±ç­ç¥å¯è½ä¼äº§çä¸åçç»æï¼å æ­¤å¯è§£éæ§å¾éè¦ã
bins = int(np.sqrt(len(df)))
#Visualization for votes
#ç¥¨æ°å¯è§å
plt.figure(figsize = (12,8))
df['Aggregate rating'].hist(bins = bins,edgecolor = 'black')
plt.title('Distribution of ratings')
plt.xlabel('Aggregate rating')
plt.ylabel('frequency')
plt.grid(False)
plt.show()
```

```python
#Calculate the most common rating range
#è®¡ç®æ®éè¯åèå´
rating_range = pd.cut(df['Aggregate rating'],bins = bins).value_counts()
most_common_rating_range = rating_range.idxmax()
print(f'the most common rating range is:{most_common_rating_range}')
```

```python
# Calculate the average vote
#è®¡ç®å¹³åç¥¨æ°
votes_count = df['Votes'].mean()
print(f'The average vote is/å¹³åç¥¨æ°:{votes_count}')
```
